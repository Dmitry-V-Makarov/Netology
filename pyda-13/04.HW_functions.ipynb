```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def doc_find():
    """
    Узнать владельца по номеру документа
    """

    inp = input('Введите номер документа: ')

    found = False
    for i in documents:
        if inp in i.values(): 
            print('Владелец документа: ' + i['name'])
            found = True

    if found == False:
        print('Документ не найден')
        
def shelf_find():
    """
    Узнать на какой полке хранится документ по его номеру
    """
        
    inp = input('Введите номер документа: ')
        
    found = False      
    for k, v in directories.items():
        if inp in v: 
            print('Документ хранится на полке: ' + k)
            found = True

    if found == False:
        print('Документ не найден')

def full_info():
    """
    Узнать на какой полке хранится документ по его номеру
    """
    for i in documents:
            for k, v in directories.items():
                if i['number'] in v:
                    print('№: ' + i['number'] + ', тип: ' + i['type'] + ', владелец: ' + i['name'] + ', полка хранения: ' + k)

def add_shelf():
    """
    Добавить новую полку
    """
    inp = input('Введите номер полки: ')
        
    if inp in directories:
        print('Такая полка уже существует. ', end = '')
    else:
        directories[inp] = []
        print('Полка добавлена. ', end = '')

    print('Текущий перечень полок: ', end = '')
    print(*list(directories.keys()), sep = ", ")

def del_shelf():
    inp = input('Введите номер полки: ')
        
    if inp in directories:
        if len(directories[inp]) != 0:
            print('На полке есть документы, удалите их перед удалением полки. ', end = '')
        else:
            del(directories[inp])
            print('Полка удалена. ', end = '')
    else:
        print('Такой полки не существует. ', end = '')

    print('Текущий перечень полок: ', end = '')
    print(*list(directories.keys()), sep = ", ")

def quit():
    print('Выход')
    cont = False

commands = {
    'p': doc_find,
    's': shelf_find,
    'l': full_info,
    'as': add_shelf,
    'ds': del_shelf,
    'q': quit
}
    
def main():
    
    while True:
        try:
            user_input = input('Введите команду: ')
            commands[user_input]()
        
            if user_input == 'q':
                break
        
        except:
            print('Ошибка ввода')

main()


    
```
