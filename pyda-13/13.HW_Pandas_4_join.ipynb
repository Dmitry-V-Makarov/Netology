### Задание 3

Дано: таблица визитов, покупок и координат по одному пользователю. Имеются дубликаты по дате и как следствие разные источники, визиты, покупки и координаты для одной и той же даты.

```python
visits_orders = pd.DataFrame(
    {'date': ['2019-11-01', '2019-11-01', '2019-11-02', '2019-11-02', '2019-11-03'], 
     'source': ['organic', 'paid', 'organic', 'paid', 'organic'], 
     'visits': [16825, 1952, 21890, 376, 19509],
     'orders': [198, 225, 201, 203, 204],
     'lat': [40.30403, 77.20309, 32.30432, 25.20393, 36.30493],
     'long': [75.30495, 34.30905, 75.40953, 49.60795, 56.00124]
    }
)

visits_orders
```

Возможное решение:
1. источник можно преобразовать в столбцы с помощью сводных таблиц
1. визиты и заказы просуммировать для каждой даты
1. широту и долготу при группировке по дате можно вносить в список. Например в ячейках lat/long за 2019-11-01 будут соответственно [40.30403, 77.20309] и [75.30495, 34.30905]. Затем для целей аналитики их можно попарно извлекать из этих списков.

Возможное использование координат:
1. в целом для всех пользователей: понимание кластеров, в которых чаще всего совершаются покупки, например, для оптимизации снабжения и доставки
1. для каждого конкретного пользователя: понимание наиболее типичного/частого места покупки, возможность улучшения рекомендательной системы
