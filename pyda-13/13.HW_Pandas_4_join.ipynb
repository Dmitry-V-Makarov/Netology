```python
ratings = pd.read_csv(r'C:\Users\dmvma\Downloads\ml-latest\ratings.csv')
ratings.head()
```

```python
ratings_f = ratings.groupby('userId').filter(lambda x: len(x) > 100)
ratings_f.head()
```

```python
ratings_agg = ratings_f.groupby('userId').agg([min, max])
ratings_agg.head()
```

```python
ratings_agg['diff'] = ratings_agg['timestamp']['max'] - ratings_agg['timestamp']['min']
ratings_agg.head()
```
