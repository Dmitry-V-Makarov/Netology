### Уровень 0

**Задание 1**. Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
mu_1, mu_2, mu_3, mu_4, mu_5 = 0, 0, 1, 1, 2
var_1, var_2, var_3, var_4, var_5 = 1, 1.5, 3, 4, 3

sigma_1 = math.sqrt(var_1)
sigma_2 = math.sqrt(var_2)
sigma_3 = math.sqrt(var_3)
sigma_4 = math.sqrt(var_4)
sigma_5 = math.sqrt(var_5)

x = np.linspace(-3, 5, 100)
plt.figure(figsize=(15,8))

plt.plot(x, stats.norm.pdf(x, mu_1, sigma_1))
plt.plot(x, stats.norm.pdf(x, mu_2, sigma_2))
plt.plot(x, stats.norm.pdf(x, mu_3, sigma_3))
plt.plot(x, stats.norm.pdf(x, mu_4, sigma_4))
plt.plot(x, stats.norm.pdf(x, mu_5, sigma_5))

plt.title('Normal Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')

plt.show()
```

```python
lambd_1, lambd_2, lambd_3, lambd_4, lambd_5 = 0.5, 1, 1.2, 2, 3

x = np.arange(0, 17, 0.1)

y_1 = lambd_1 * np.exp(-lambd_1 * x)
y_2 = lambd_2 * np.exp(-lambd_2 * x)
y_3 = lambd_3 * np.exp(-lambd_3 * x)
y_4 = lambd_4 * np.exp(-lambd_4 * x)
y_5 = lambd_5 * np.exp(-lambd_5 * x)

plt.figure(figsize=(15,8))

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.plot(x, y_4)
plt.plot(x, y_5)

plt.title('Exponential Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

```python
rv_1 = stats.t(df=30, loc=0, scale=1)
x_1 = np.linspace(rv_1.ppf(0.0001), rv_1.ppf(0.9999), 100)
y_1 = rv_1.pdf(x) 

rv_2 = stats.t(df=10, loc=1, scale=1)
x_2 = np.linspace(rv_2.ppf(0.0001), rv_2.ppf(0.9999), 100)
y_2 = rv_2.pdf(x) 

rv_3 = stats.t(df=15, loc=0.5, scale=1)
x_3 = np.linspace(rv_3.ppf(0.0001), rv_3.ppf(0.9999), 100)
y_3 = rv_3.pdf(x) 

rv_4 = stats.t(df=40, loc=0, scale=1)
x_4 = np.linspace(rv_4.ppf(0.0001), rv_4.ppf(0.9999), 100)
y_4 = rv_4.pdf(x) 

rv_5 = stats.t(df=50, loc=1.2, scale=1)
x_5 = np.linspace(rv_5.ppf(0.0001), rv_5.ppf(0.9999), 100)
y_5 = rv_5.pdf(x) 

plt.figure(figsize=(15,8))

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.plot(x_3,y_3)
plt.plot(x_4,y_4)
plt.plot(x_5,y_5)

plt.title('Student t-Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

**Задание 3**. Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
p_red = 54/(54+4+28)
p_orange = 4/(54+4+28)
p_green = 28/(54+4+28)

I_red = math.log2(1/p_red)
I_orange = math.log2(1/p_orange)
I_green = math.log2(1/p_green)

print(I_red, I_orange, I_green)
```

```python
p_red = 28/(28*3)
p_orange = 28/(28*3)
p_green = 28/(28*3)

I_red = math.log2(1/p_red)
I_orange = math.log2(1/p_orange)
I_green = math.log2(1/p_green)

print(I_red, I_orange, I_green)
```
