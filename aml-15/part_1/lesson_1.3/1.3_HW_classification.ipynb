```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
```

```python
df.head()
```

#### Basic statistics and preprocessing

```python
df.describe()
```

```python
# Replace ? with NaN
df[df == '?'] = np.nan
```

```python
# Basic stats after dropping NaN
df.dropna().describe()
```

```python
# statistics stay roughly the same after dropping rows with NaN
```

```python
# % of missing values per column
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# definitely dropping all NaN rows
df = df.dropna()
```

#### Dealing with categorical features

```python
df_numerical = pd.get_dummies(df, drop_first=True)
```

```python
df_numerical.head()
```

```python
y = df_numerical['income_>50K'] # 1 means > 50K and 0 means <= 50K
```

```python
X = df_numerical.drop('income_>50K', axis = 1)
```

#### Splitting, scaling and modelling data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
# Logistic Regression pipeline
```

```python
# Pipeline
steps = [('scaler', StandardScaler()),
         ('lr', LogisticRegression(max_iter=10000))]

pipeline = Pipeline(steps)

# Hyperparameter space
c_space = np.logspace(-5, 8, 15)
param_grid = {'lr__C': c_space}

# GridSearchCV
cv_lr = GridSearchCV(pipeline, param_grid, cv=5)

# Fit to the training set
cv_lr.fit(X_train,y_train)

# Predict the labels of the test set: y_pred
y_pred_lr = cv_lr.predict(X_test)
```

```python
# Compute and print metrics
print("Accuracy: {}".format(cv_lr.score(X_test, y_test))) # accuracy
print(classification_report(y_test, y_pred_lr))
print("Tuned Model Parameters: {}".format(cv_lr.best_params_))
```
